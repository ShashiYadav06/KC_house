### IMPORTING NECESSARY MODULES AND MODELS

```python
data_set=dumm.copy()
```

```python
from sklearn.model_selection import KFold , train_test_split ,cross_val_score
from sklearn.preprocessing import StandardScaler , LabelEncoder , OneHotEncoder

from sklearn.linear_model import LinearRegression , Ridge , Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR #support vector classifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import r2_score
from sklearn.feature_selection import SelectFromModel
```

```python
#spliting data into test and train
X=data_set.drop("price",axis=1).values
y=data_set["price"].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=7)
```

```python
print("Train",X_train.shape,y_train.shape)
print("Test",X_test.shape,y_test.shape)
```

```python
pipelines = []
seed =2

pipelines.append(
                ("Scaled_Rigde",
                Pipeline([
                    ("Scalar",StandardScaler()),
                    ("Rigde",Ridge(random_state=seed,tol=10))
                    ]))
                )
pipelines.append(
                ("Scaled_Lasso",
                Pipeline([
                    ("Scalar",StandardScaler()),
                    ("Lasso",Lasso(random_state=seed,tol=1))
                    ]))
                )
pipelines.append(
                ("Scaled_RandomForest",
                Pipeline([
                    ("Scalar",StandardScaler()),
                    ("RandomForest",RandomForestRegressor(random_state=seed))
                    ]))
                )
pipelines.append(
                ("Scaled_SVR",
                 Pipeline([
                     ("Scaler", StandardScaler()),
                     ("SVR",  SVR(kernel='linear', C=1e2, degree=5))
                 ])
                )
                )
pipelines.append(
                ("Scaled_DT",
                Pipeline([
                    ("Scalar",StandardScaler()),
                    ("Decision_Tree",DecisionTreeRegressor())
                    ]))
                )
pipelines.append(
                ("Scaled_Knn",
                Pipeline([
                    ("Scaler",StandardScaler()),
                    ("Scaled_Knn",KNeighborsRegressor())
                    ]))
                )

scoring="r2"
n_folds=7

results , names = [],[]

for name,model in pipelines:
    kfold=KFold(n_splits=n_folds,random_state=seed)
    cv_result=cross_val_score(model,X_train,y_train,
                              scoring=scoring,cv=kfold,n_jobs=-1)
    names.append(name)
    results.append(cv_result)

    msg = "%s: %f (+/- %f)" % (name, cv_result.mean(),  cv_result.std())
    print(msg)
```

```python
fig = plt.figure(figsize=(15,6))
fig.suptitle('Algorithm Comparison', fontsize=22)
ax = fig.add_subplot(111)
sns.boxplot(x=names, y=results)
ax.set_xticklabels(names)
ax.set_xlabel("Algorithmn Name", fontsize=20)
ax.set_ylabel("R Squared Score of Models", fontsize=18)
ax.set_xticklabels(ax.get_xticklabels(),rotation=0)
plt.show()
```
